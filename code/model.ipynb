```python
import pandas as pd

df_raw = pd.read_csv('../data/dados.csv')
```

```python
df = df_raw
```

```python
df.fillna({'genero': 'Não informado', 'descricao': 'Não informado', 'autor': 'Não informado','idioma': 'Não informado','editora': 'Não informado'}, inplace=True)
```

```python
df.drop(columns=['ISBN_10','ISBN_13'], inplace=True)
```

```python
df['genero'] = df['genero'].apply(lambda x: x.split('/') if isinstance(x, str) else [])
```

```python
df.head()
```

```python
import category_encoders as ce

# Unir os gêneros em uma única string por linha, ex: 'ação,comédia'
df['genero'] = df['genero'].apply(lambda x: ','.join(sorted(x)) if isinstance(x, list) else x)

# Inicializar o codificador binário
encoder = ce.BinaryEncoder(cols=['genero'], drop_invariant=True)

# Aplicar o codificador ao DataFrame
df_encoded = encoder.fit_transform(df)

df = df_encoded
```
